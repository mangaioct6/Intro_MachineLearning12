## 2. Create a new dataframe using week_11/temperatures/temperature.csv file from last class. It should have the datetime column and one of the cities. Create a new boolean column called is_freezing that indicates if the temperature is at or below freezing. Use a lambda to create this column. Is there a better way to do this than with a lambda?

```python
temp_df=pd.read_csv('../Time-Series-Analysis-11/temperatures/temperature.csv',parse_dates=['datetime'],index_col='datetime')
portland_df=pd.DataFrame(temp_df.iloc[:,1])
portland_df.head()
```

```python
portland_df.dropna()
```

```python
portland_df['is_freezing'] = portland_df['Portland'].map(lambda n:True if n<=273.15 else False)
portland_df
```

### Is there a better way to do this than with a lambda?

* For this question we felt lambda is the better way. Because it is **simple and straightforward function** that requires only the argument and condition and lambda keyword.

* No iterating loops inside dataframe

* saves time and works efficiently. 

## BONUS: if you still have time, try to do a pivot of the data in question 2 so you have the hourly data for all the cities and a flag for if it's freezing

```python
#.stack() helps us to reshape the dataframe
stack=temp_df.stack()
```

```python
stack_df=pd.DataFrame(stack)
stack_df.columns=['temp_k']
stack_df['Flag_Col']= stack_df['temp_k'].map(lambda n:True if n<=273.15 else False)
stack_df
```
